```python
import bs4
import requests
import lxml
```

```python
def get_fighter_a(fight_url):
    
    #Checks that the url is from www.mmadecisions.com
    
    if 'http://www.mmadecisions.com' in fight_url:
        
        #Checks that the url corresponds to a fight rather than e.g. an event or a judge
        
        if 'http://www.mmadecisions.com/decision' in fight_url:
    
            #Generates a BeautifulSoup object for the url
            
            get_fight = requests.get(fight_url)
            get_fight_soup = bs4.BeautifulSoup(get_fight.text, 'lxml')

            #Selects the text from the class 'decision-top'
            
            fighter_a = get_fight_soup.select('.decision-top')[0].text
            
            #Formats the text to just show the fighter name
            
            fighter_a = fighter_a.replace('\n','').replace('\r','').replace('\t','')

            print(fighter_a)
        
        else:
            print('Error: you must enter the url for a specific fight')
    
    else:
        print('Error: you must enter a url from www.mmadecisions.com')
```
